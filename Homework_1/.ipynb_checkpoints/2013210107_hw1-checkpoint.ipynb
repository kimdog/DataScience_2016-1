```python
def get_movie_detaill_page(movie_id):
    print "fetching " , movie_id
    
    myR = requests.get('http://www.imdb.com/title/'+movie_id)
    return myR.text
```

```python
from pattern import web
```

```python
def parsing_detail_page(html):
    
    element = web.Element(html)
    
    duration = element('.subtext time')[0].content
    
    genre = []
    for i in element('.subtext a span'):
        sample_content = i.content
        genre.append(sample_content)
    #genre  = element('.subtext a span')
   
    rating = element('.ratingValue strong span')[0].content
    rating_count = element('.imdbRating a span')[0].content
    
    review_count = element('.titleReviewBarItem.titleReviewbarItemBorder span a')[0].content
    
    
    title = element('.title_wrapper h1')[0].content
    
    year = element('.title_wrapper h1 a')[0].content
    
    
    
    dict = {
     'duration': duration,
     'genre': genre,
     'rating': rating,
     'rating_count': rating_count,
     'review_count': review_count,
     'title': title,
     'year': year
       }

    return dict
```

```python
import re
```

```python
def parsing_convert_dt(data):
    
    #duration
    duration = data.get('duration')
    cp = re.findall('[0-9]+',duration)
    hour = int(cp[0])
    min = int(cp[1])
    duration = hour*60 + min
    
    #genre
    genre = data.get('genre')
    
    #rating
    rating = data.get('rating')
    rating = float(rating)
    
    #rating_count
    rating_count = data.get('rating_count')
    cp = re.findall('[0-9]+',rating_count)
    rating_count = ''
    for i in range(len(cp)) :
        rating_count = rating_count + cp[i]
        
    rating_count = int(rating_count)
    
    #review_count
    review_count = data.get('review_count')
    cp = re.findall('[0-9]+',review_count)
    
    review_count = ''
    for i in range(len(cp)) :
        review_count = review_count + cp[i]
    
    review_count = int(review_count)
    
    #title
    title = data.get('title')
    cp = re.split('&nbsp',title)
    title = cp[0]
    
    #year
    year = data.get('year')
    year = int(year)
    
    #result
    dict = {
     'duration': duration,
     'genre': genre,
     'rating': rating,
     'rating_count': rating_count,
     'review_count': review_count,
     'title': title,
     'year': year
       }
    
    return dict
```

```python
# each page has 50 movies sorted by popularity
def get_movie_ids_in_year(year, page=1): # if no parameters are passed, set default as page 1  
    
    page = (page - 1) * 50 + 1
    html = "http://www.imdb.com/search/title?title_type=feature&release_date=" + str(year)
    
    if page > 1 :
        html = html + "&start=" + str(page)
        
    myR_2 = requests.get(html)
    re_html = myR_2.text
    
    element_2 = web.Element(re_html)
    
    result_array = []
    
    for i in range(50):
        sampleText = element_2('div#main > table tr td.image a')[i].source
        cp = re.split('/',sampleText)    #cp[2] is our target id data
        data = cp[2]
        result_array.append(data)
    
    return result_array
```

```python
#del data['tt2091935']
def origin_to_clean(origin_data):

    toDelete_data_movies = [ movie_id 
               for movie_id, movie_data in origin_data.items()
               if movie_data is None]

    for del_Data in toDelete_data_movies:
        del origin_data[del_Data]

    return origin_data
```

```python
clean_data = origin_to_clean(data)

empty_data_movies = [ movie_id 
               for movie_id, movie_data in clean_data.items()
               if movie_data is None]

print empty_data_movies
```

```python
import pandas as pd

movies = pd.DataFrame(clean_data)
movies.head()
```

```python
tmp_df = movies.copy().transpose()
tmp_df['duration'].isnull().head()
```

```python
sum_of_duration = 0
Num = 0

#get min_duration
for duration in tmp_df['duration']:
    if(duration != None):
        sum_of_duration += duration
        Num += 1

min_duration = sum_of_duration / Num
print sum_of_duration, Num, min_duration

#update None values
Num = 0
for duration in tmp_df['duration']:
    if(duration == None):
        tmp_df['duration'][Num] = min_duration
        #print tmp_df['duration'][Num]
    Num += 1
```

```python
movies = tmp_df
year_group_movies = movies.groupby('year')
#movies.describe()
year_group_movies.describe()
```
